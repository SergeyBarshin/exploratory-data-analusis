### Demonstration of the main Pandas methods
Основые структуры данных в pandas: **Series** и **DataFrames**.

- **Series**  одномерный массив одного типа данных 
- **DataFrame** двумерная таблица в которой каждый столбец содержит данные одного типа
  
По сути **DataFrame** является набором **Series**.

### Индексация и извлечение данных
Чтобы извлечь одну колонку: `DataFrame['Name']`
  
**Boolean indexing**: `df[P(df['Name'])]`, где `P` некоторое логическое условие, которое проверится для каждого элемента колонки `Name`

### Применение функций к ячейкам, столбцам и строкам

Чтобы примениьт некоторую функцию к кадой колонке `apply()`:

```python
df.apply(np.max)
```

Чтобы применить для каждой строки. Лямбда-функции очень удобны в таких сценариях.

```python
df[df["State"].apply(lambda state: state[0] == "W")].head() # используем лямбду для булевой индексации
```

`map` может быть использован, чтобы заменить значения в колонкуе передавая словарь вида `{old_value: new_value}` как аргумент

Тоже самое можно сделать при помощи `replace`.

Разница: `replace` ничего не сделает, для значений которые не нашел, `map` заменит изи на `NaN`




```python
d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)

df = df.replace({"Voice mail plan": d})

df.head()
```

### Grouping
В целом, группировка данных в Pandas работает следующим образом:

`df.groupby(by=grouping_columns)[columns_to_show].function()`

1. `groupby` делит grouping_columns на их значения. Они становятся новым индексом в результируютом df.
2. Затем выбираются интересующие столбцы (`columns_to_show`). Если `columns_to_show` не включены, будут включены все предложения, не связанные с группировкой.
3. Наконец, одна или несколько функций применяются к полученным группам на выбранные столбцы.


```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

### Summary tables

Предположим, мы хотим увидеть, как в нашем датасете распределены две переменные - `Churn` и `International plan`. Для этого мы можем построить `contingency table`, используя метод `crosstab`:

```python
print(df.columns.duplicated())

pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

Мы видим, что большинство пользователей лояльны и не пользуются дополнительными услугами (Международный план/Голосовая почта).

`pivot_table` принимает следующие параметры:
- `values` – список переменных, для которых считаем статистики,
- `index` – список переменных по которым групируем,
- `aggfunc` – какие статистики нужно псчитать sum, mean, maximum, minimum или что-то другое.

```python
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```

### DataFrame transformations
В pandas можно добавлять колонки многими способами.

```python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
# loc parameter is the number of columns after which to insert the Series object
# we set it to len(df.columns) to paste it at the very end of the dataframe
df.head()
```

Можно легче добавить столбец без создания промежуточного экземпляра фичи

```python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
```

Для удаления колонок и строк, используется метод `drop`

Обязательный аргумент `axis`. `axis=1` - удаляем колонку; `axis=2` - строку.

Аргумент `inplace` , следует ли изменять исходную df. С `inplace=False` метод не изменяет существующий DataFrame и возвращает новый. С `inplace=True` он изменяет df.

```python
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
df.drop([1, 2]).head()
```
